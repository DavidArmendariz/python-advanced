```python
import math
class Complex:
    # Part 1
    def __init__(self, re, im):
        self.re = re
        self.im = im

    def __str__(self):
        if self.im < 0:
            return f"{self.re}{self.im}i"
        else:
            return f"{self.re}+{self.im}i"
    
    # Part 2
    def __add__(self, other):
        if isinstance(other, int) or isinstance(other, float):
            self.re += other
        elif isinstance(other, Complex):
            self.re += other.re
            self.im += other.im
        else:
            raise TypeError
    
    def __sub__(self, other):
        if isinstance(other, int) or isinstance(other, float):
            self.re -= other
        elif isinstance(other, Complex):
            self.re -= other.re
            self.im -= other.im
        else:
            raise TypeError

    # Part 3
    def __mul__(self, other):
        if isinstance(other, int) or isinstance(other, float):
            self.re *= self.other
            self.im *= self.other
        elif isinstance(other, Complex):
            self.re = self.re*other.re-self.im*other.re
            self.im = self.re*other.im+self.im*other.re
        else:
            raise TypeError
    
    def __truediv__(self, other):
        pass

    # Part 4
    def __eq__(self, other):
        if isinstance(other, int) or isinstance(other, float):
            if self.re == other.re and self.im == 0:
                return True
            else:
                return False
        elif isinstance(other, Complex):
            if self.re == other.re and self.im == other.im:
                return True
            else:
                return False
        else:
            raise TypeError
    
    def __abs__(self):
        return math.sqrt(self.re**2+self.im**2)
```
